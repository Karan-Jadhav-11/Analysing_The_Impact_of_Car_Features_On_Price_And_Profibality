```python
import pandas as pd

def find_outliers_iqr(data, column):
  """
  Identifies outliers in a specific column of a DataFrame using the IQR method.

  Args:
      data: pandas DataFrame.
      column: str, the name of the column to analyze.

  Returns:
      pandas Series containing the outliers.
  """
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  print("IQR : ", IQR)
  print("Lower Bound:", lower_bound)
  print("Upper Bound:", upper_bound)
  outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)][column]
  return outliers

# Assuming df_num1 is your DataFrame with numerical columns
# Example usage:
outliers_hp = find_outliers_iqr(df_num1, 'Engine HP')
print("Outliers in Engine HP column:", outliers_hp)

# You can repeat this for other numerical columns in df_num1
for column in df_num1.columns:
  outliers = find_outliers_iqr(df_num1, column)
  if not outliers.empty:
    print(f"Outliers in {column}:", outliers)
outl=outliers
```

```python
!pip install scipy
import pandas as pd
from scipy.stats.mstats import winsorize

def winsorize_outliers(df, column_list, limits=(0.01, 0.05)):
  """
  Winsorizes outliers in specified columns of a DataFrame.

  Args:
      df: pandas DataFrame.
      column_list: list of str, the names of the columns to winsorize.
      limits: tuple, lower and upper percentile limits for winsorization.
              Defaults to (0.05, 0.05), which means the bottom 5% and
              top 5% of values will be winsorized.

  Returns:
      pandas DataFrame with winsorized columns.
  """
  for column in column_list:
    df[column] = winsorize(df[column], limits=limits)
  return df

# Assuming df_num1 is your DataFrame with numerical columns
# and numerical is a list of your numerical column names

# Winsorize the outliers in numerical columns
df_num1_winsorized = winsorize_outliers(df_num1.copy(), numerical)
# We use .copy() to avoid modifying the original df_num1

print("DataFrame with winsorized outliers:")
print(df_num1_winsorized.head()) # Display the first few rows to verify
```
